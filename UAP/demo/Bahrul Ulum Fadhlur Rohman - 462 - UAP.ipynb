```python
# 1. Sequence
data_tuple = ('Pensil', 1500, 'Buku', 5000, 'Penggaris', 2000)

def pisah_data(data):
  return list(data[::2]), list(data[1::2])

def buat_dict(data_barang, data_harga):
    return dict(zip(data_barang, data_harga))

print(data_tuple)

list_barang, list_harga = pisah_data(data_tuple)
print(list_barang, list_harga)

dic_result = buat_dict(list_barang, list_harga)
print(dic_result)
```

```python
# 2. Function
def isPalindrome(data):
    balik_str = ''.join(filter(str.isalnum, data.lower()))
    return balik_str == balik_str[::-1]

string = 'radar'
result_str = isPalindrome(string)
print(result_str)
```

```python
# 3. Pure Function
def tambah_angka(angka, total):
  return angka + total

total = 0;
hasil = tambah_angka(5, total)
print (hasil)
```

```python
# 4. Lambda
klasifikasi_angka = lambda x: "Positif" if x > 0 else "Negatif" if x < 0 else "Nol"
pencarian_angka = lambda list_angka, find: "Found" if find in list_angka else "Not Found"

angka = -3
result = klasifikasi_angka(angka)
print(result)

ex_angka = [1,2,3,4,5]
cari = 5
result_cari = pencarian_angka(ex_angka, cari)
print(result_cari)
```

```python
# 5. List Comprehension
ganjil = [i for i in range (1, 50) if i % 2 != 0]

print(ganjil)
```

```python
# 6. Generator
def generator():
  for i in range (1, 50):
    if i % 2:
      yield i

hasil = generator()
for i in hasil:
  print(i, end=", ")
```

```python
# 7. HOF (NIM Genap => return value )
def sisagold():
    return lambda a, b: a - b

result = sisagold()
output = result(10, 3)
print(output)
```

```python
# 8. Filtering
def isPalindrome(data):
    balik_str = ''.join(filter(str.isalnum, data.lower()))
    return balik_str == balik_str[::-1]

string = 'Ada apa dengan si Bahrul Ulum Fadhlur Rohman'
kalimat = string.split()

palindrome = list(filter(lambda data: isPalindrome(data), kalimat))
no_palindrome = list(filter(lambda data: not isPalindrome(data), kalimat))

print('data = ', kalimat)
print('polindrome = ', palindrome)
print('non-polindrome = ', no_palindrome)
```

```python
# 10. Reduce
from functools import reduce

data_angka = [3, 9, 15, 21, 27, 33, 39, 45]
result = reduce(lambda x, y: x + y, data_angka)
print(result)
```

```python
# 11. Closure
def hitung_pangkat(pangkat):
  def hitung_angka(angka):
    return angka ** pangkat
  return hitung_angka

result = hitung_pangkat(2)
print(result(5))
```

```python
# 12. Decorator
def convert_to_sentence(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        if result:
            return "Yup, it's a palindrome!"
        else:
            return "Nope, it's not a palindrome."
    return wrapper

@convert_to_sentence
def isPalindrome_decorated(data):
    balik_str = ''.join(filter(str.isalnum, data.lower()))
    return balik_str == balik_str[::-1]

result = isPalindrome_decorated("sus")
print(result)
```

```python
# 13. Data Visualization
import matplotlib.pyplot as plt

data = [
    ('Bus', 5, 200),
    ('Trem', 8, 150),
    ('Kereta', 12, 300),
    ('Minibus', 6, 180),
    ('Tram', 9, 220),
]

jenis_kendaraan = list(map(lambda x: x[0], data))
penggunaan_energi = list(map(lambda x: x[1], data))
biaya_operasional = list(map(lambda x: x[2], data))

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.bar(jenis_kendaraan, penggunaan_energi, color='skyblue')
plt.title("penggunaan_energi")
plt.xlabel("jenis_kendaraan")
plt.ylabel("penggunaan_energi")

plt.subplot(1, 3, 2)
plt.bar(jenis_kendaraan, biaya_operasional, color='skyblue')
plt.title("biaya_operasional")
plt.xlabel("jenis_kendaraan")
plt.ylabel("biaya_operasional")

plt.subplot(1, 3, 3)
plt.scatter(penggunaan_energi, biaya_operasional, color='darkblue', label='jenis_kendaraan')
plt.title("Hubungan Biaya dan Energi")
plt.xlabel("penggunaan_energi")
plt.ylabel("biaya_operasiona;")
plt.legend(loc="upper left")

plt.tight_layout()
plt.show()
```
